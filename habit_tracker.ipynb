```python
# HABIT TRACKER

def habit_tracker():
    habits = []
    completed = []
    progress_list = []

    try:
        with open("habits.txt", "r") as f:
            for line in f:
                habits.append(line.strip())
    except FileNotFoundError:
        pass

    try:
        with open("completed.txt", "r") as f:
            for line in f:
                completed.append(line.strip())
    except FileNotFoundError:
        pass

    try:
        with open("progress.txt", "r") as f:
            for line in f:
                progress_list.append(float(line.strip()))
    except FileNotFoundError:
        pass

    def welcome():
        print('''                      Menu
        1. List of habits
        2. Add a habit
        3. Mark habit as completed
        4. Show today's progress
        5. Show weekly summary
        6. Exit''')
    
    def add_a_habit(habit_name):
        if habit_name in habits:
            print("This habit already exists!")
            return
        habits.append(habit_name)
        print("New habit added successfully.")
    
    def mark_habit_as_completed(completed_habit_name):
        if completed_habit_name in completed:
            print("You have already mentioned it!, complete another task")
            return
        completed.append(completed_habit_name)
        print("Good job!")
    
    def show_today_progress():
        if len(habits) == 0:
            print("There is no any habit added yet! Please add a habit.\n")
            return
        else:
            progress = (len(completed)/len(habits))*100
            progress_list.append(progress)
            # Save progress list
            with open("progress.txt", "w") as f:
                for p in progress_list:
                    f.write(str(p) + "\n")
    
            print("Today's Progress:", progress, "%")
            if progress >= 80:
                print("Excellent, you are close!")
    
            elif progress >= 50 and progress < 80:
                print("Good, keep doing!")
    
            else:
                print("Try to improve")
    
            progress_list.append(80)
    
    def weekly_summary():
        print(progress_list)
    
    print("                  Habit Tracker")
    
    welcome()
    
    while True:
        i = int(input("Enter the respective no. to perform a task.\n"))
        if i == 1:
            print("Here is your habits list")
            print(habits) 
        elif i == 2:
            u = input("Write a new habit to be be better of yourself.\n")
            add_a_habit(u)
            print("Your habit list: ", habits)
            print("\n")
            welcome()
            
        elif i == 3:
            v = input("Type which habit have you completed.\n")
            mark_habit_as_completed(v)
            print("\n")
            welcome()
        
        elif i == 4:
            show_today_progress()
            completed.clear()
            print("\n")
            welcome()
        
        elif i == 5:
            weekly_summary()
        
        elif i == 6:
            print("Goodbye!")
            break
    
        else:
            print("Invalid choice!, Try again")

       


habit_tracker()
```
